#Практическое задание #1

```python

class Chocolate:

    def __init__(self, price: float, additions: str = None):
        if price > 0:
            self.price = price
        else:
            raise ValueError('Дешево - это хорошо, но не на столько')

        self.addtions = additions

    def show_my_chocolate(self):
        if self.addtions != None:
            print(f'Шоколадный батончик, добавка: {self.addtions}')
        else:
            print('Обычный шоколадный батончик')

    def __str__(self):
        print(f'Шоколадный батончик, {self.price} рублей')

choco1 = Chocolate(120.7)
choco2 = Chocolate(300, 'фундук')

choco1.__str__()
choco2.show_my_chocolate()
```
