Функция для поиска короткого имени

```python
def name_finder(full_name):
    #contractions = ['Mrs.', 'Mr.', 'Miss.', 'Master.', 'Rev.', 'Dr.', 'Col.', 'Sir.', 'Major.']
    #for i in contractions:
    #if i in full_name:
    if 'Mrs.' not in full_name:
        name_start = full_name.find('.')+2
        if full_name[name_start:].find(' ') != -1:
            name_end = full_name[name_start:].find(' ')+name_start
            return full_name[name_start:name_end]
        else:
            return full_name[name_start:]
    else:
        name_start = full_name.find('(') + 1
        if full_name[name_start:].find(' ') != -1:
            name_end = full_name[name_start:].find(' ')+name_start
            return full_name[name_start:name_end]
        else:
            return full_name[name_start:]
```
