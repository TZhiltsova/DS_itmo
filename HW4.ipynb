```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
passengers = pd.read_csv('train.csv')

x = passengers['Survived'].apply(lambda x: str(x)).unique()
y = passengers['Survived'].value_counts()
fig, ax = plt.subplots()
ax.bar(x=x, height=y)
```

```python
x = passengers['Pclass'].apply(lambda x: str(x)).unique()
y = passengers['Pclass'].value_counts()
fig, ax = plt.subplots()
ax.bar(x=x, height=y)
```

```python
x = passengers['Age'].dropna().apply(lambda x: str(x)).unique()
y = passengers['Age'].dropna().value_counts()
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = passengers['Sex'].apply(lambda x: str(x)).unique()
y = passengers['Sex'].value_counts()
fig, ax = plt.subplots()
ax.bar(x=x, height=y)
```

```python
x = passengers['Parch'].apply(lambda x: str(x)).unique()
y = passengers['Parch'].value_counts()
fig, ax = plt.subplots()
ax.bar(x=x, height=y)
```

```python
labels= passengers['Survived'].apply(lambda x: str(x)).unique()
value = passengers['Survived'].value_counts()
fig, ax = plt.subplots()
ax.pie(value, labels=labels, autopct= ' %1.1f%%');
```

```python
values = passengers['Pclass'].value_counts().tolist()
values.append(passengers['Pclass'].sum())
labels= passengers['Pclass'].apply(lambda x: str(x)).unique().tolist()

data = dict(
    categories = ['3', '1', '2', 'Total'],
    parent = ['', 'Total', 'Total', 'Total'],
    value = values
)

fig = px.sunburst(data, names='categories', values = 'value', branchvalues='total')
fig.show()
```
